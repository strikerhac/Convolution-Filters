```python
import numpy
import matplotlib
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt 
import imageio
```

```python
#im = Image.open("two_cats.jpg").convert("L").save("cats.jpg")
```

```python
def padding(container,filter_size):
    pad_count = np.int(np.floor(filter_size/2))
    a = np.zeros((1,container.shape[1]))
    for up in range(pad_count):
        container = np.concatenate((container,a),axis=0)
    for down in range(pad_count):
        container = np.concatenate((a,container),axis=0)
    b = np.zeros((container.shape[0],1))
    for left in range(pad_count):
        container = np.concatenate((container,b),axis=1)
    for right in range(pad_count):
        container = np.concatenate((b,container),axis=1)
    return container
```

```python
def convolve(image,kernel,size):
    List = []
    sum = 0
    for shift_r in range(image.shape[0]-(size-1)):
        for shift_c in range(image.shape[1]-(size-1)):
            for r in range(len(kernel)):
                for c in range(len(kernel[r])):
                    sum = sum + (image[r+shift_r][c+shift_c] * kernel[r][c])
            List.append(sum/(size*size))    
            sum = 0
    return List
```

```python
def average_blur(image_url,size):
    #Filter construction
    avg_filter = np.ones((size,size))
    #read image and pad image
    container = imageio.imread(image_url)
    image = padding(container,size)
    #convolve padded image
    List = convolve(image,avg_filter,size)
    result = (np.asarray(List)).reshape(image.shape[0]-(size-1),image.shape[1]-(size-1))
    
    return result
```

```python
def weighted_average_blur(image_url,size):
    #Filter construction
    avg_filter = np.zeros((size,size))
    center = np.floor(size/2)
    for r in range(size):
        for c in range(size):
            if (np.abs(r-center)+np.abs(c-center)) == 0:
                avg_filter[r][c] = 2
            else:
                avg_filter[r][c] = 1/(np.abs(r-center)+np.abs(c-center))
    mul_factor = 1/avg_filter[0][0]
    avg_filter = np.round(avg_filter*mul_factor)
    #read image and pad image
    container = imageio.imread(image_url)
    image = padding(container,size)
    #convolve padded image
    List = convolve(image,avg_filter,size)
    result = (np.asarray(List)).reshape(image.shape[0]-(size-1),image.shape[1]-(size-1))
    
    return result
```

```python
def gaussian_blur(image_url,size):
    #Filter construction
    s, k = 1, (np.int((size-1)/2)) #  generate a (2k+1)x(2k+1) gaussian kernel with mean=0 and sigma = s
    probs = [np.exp(-z*z/(2*s*s))/np.sqrt(2*np.pi*s*s) for z in range(-k,k+1)] 
    kernel = np.outer(probs, probs)
    #read image and pad image
    container = imageio.imread(image_url)
    image = padding(container,size)
    #convolve padded image
    List = convolve(image,kernel,size)
    result = (np.asarray(List)).reshape(image.shape[0]-(size-1),image.shape[1]-(size-1))
    
    return result
```

```python
def median_filter(image_url,size):
    #read image and pad image
    container = imageio.imread(image_url)
    image = padding(container,size)
    
    List = []
    temp = []
    #convolve padded image
    for shift_r in range(image.shape[0]-(size-1)):
        for shift_c in range(image.shape[1]-(size-1)):
            for r in range(size):
                for c in range(size):
                    temp.append(image[r+shift_r][c+shift_c])
            List.append(np.median(temp))
            temp = []
            
    result = (np.asarray(List)).reshape(image.shape[0]-(size-1),image.shape[1]-(size-1))
    return result
```

```python
def edge_detector(image_url,size):
    #Vertical Filter construction
    kernel_v = np.zeros((size,size))
    for num in range(size):
        kernel_v[0][num] = -1
        kernel_v[size-1][num] = 1
    #Horizontal Filter construction
    kernel_h = np.zeros((size,size))
    for num in range(size):
        kernel_h[num][0] = -1
        kernel_h[num][size-1] = 1
    #read image and pad image
    container = imageio.imread(image_url)
    image = padding(container,size)
    #convolve padded image
    List = convolve(image,kernel_v,size)
    result = (np.asarray(np.abs(List))).reshape(image.shape[0]-(size-1),image.shape[1]-(size-1))
    List = convolve(image,kernel_h,size)
    result = result + (np.asarray(np.abs(List))).reshape(image.shape[0]-(size-1),image.shape[1]-(size-1))
    
    return (result*9)
```

```python
def unsharpMasking(img2):
    img1 = np.asarray(img2)
    kernel = (np.ones((3, 3), np.float32) / 9)
    width, height = img2.size
    imgarr = np.asarray(img2)
    imgArr = np.zeros((height, width), np.int32)
    for i in range(1, height):
        for j in range(1, width):
            num = 0.0
            for x in range(kernel.shape[0]):
                for y in range(kernel.shape[1]):
                    num = num + (kernel[x][y] * imgarr[i - (kernel.shape[0]) + x][j - (kernel.shape[0]) + y])
            imgArr[i - 1][j - 1] = num

    imgArr = (img1 - imgArr) + img1
    plt.imshow(imgArr,cmap = matplotlib.cm.Greys_r)
```
